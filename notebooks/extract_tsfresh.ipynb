```python
import tsfresh
import json
import tqdm
```

```python
results = []
scores, datasets, models = get_global_const()
```

```python
for dataset_name in datasets:
    json_path = os.path.join(Path('data' + '/' + 'datasets_metrics' + '/' + 'datasets'), Path(dataset_name + ".json"))

    print(dataset_name)
    
    with open(json_path, 'r') as f:
        time_data = json.load(f)

    time_f = []
    for series_id, series in enumerate(time_data["X"]):
        for t, v in enumerate(series[0]):
            time_f.append({"id": series_id, "time": t, "v": v})

    df_time = pd.DataFrame(time_f)

    df_features = tsfresh.extract_features(
        df_time,
        column_id="id",
        column_sort="time",
        column_value="v",
    )

    df_mean_features = df_features.mean(axis=0)

    row_dict = {"Name": dataset_name}
    for col_name in df_mean_features.index:
        row_dict[col_name] = df_mean_features[col_name]

    results.append(row_dict)
    # print(results)
df_results = pd.DataFrame(results)



df_results.to_csv(Path("data/datasets_features/tsfresh_full_features_mean_by_series.csv"))
```
